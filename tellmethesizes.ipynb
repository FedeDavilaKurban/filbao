```python
# Read the data and random files and tell me how big they are
import numpy as np
import pandas as pd

datafile = "../data/sdss_dr72safe0_sigma_3.0.csv"
randomfile = '../data/lss_randoms_combined_cut_LARGE.csv'

# Data
zmin, zmax = .07, .2
grmin = 0.8
magmax = -21.2
data = pd.read_csv(datafile)
data = data[(data['red'] > zmin) & (data['red'] < zmax) & (data['gr'] > grmin) & (data['mag_abs_r'] < magmax)]

# Randoms
randoms = pd.read_csv(randomfile)

print(f"Main sample size: {len(data)}")
print(f"Random sample size: {len(randoms)}")
print(f"Randoms to data ratio: {len(randoms) / len(data):.2f}")
```
